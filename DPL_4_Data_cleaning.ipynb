* Keeping the data as it is
* removing that particular data
* replacing the data with dummy data (imputation)

**Data Imputation**

```python
#find mean value of 'pre_movie_score' column
df['pre_movie_score'].mean()
```

```python
df['pre_movie_score'] = df['pre_movie_score'].fillna(df['pre_movie_score'].mean())
```

```python
df
```

```python
df['age'] = df['age'].fillna(int(df['age'].mean()))
```

```python
df
```

```python
x = round(df['age'].mean())
```

```python
x
```

```python
df['age'] = df['age'].fillna(x)
```
